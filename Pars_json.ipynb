```python
def JSON_pars(file_name:str, privat_page=False):
    
    '''
    Функция для парсинга JSON файлов.
    
    На вход подается название файла,
    на выходе датасет и сохранение файла в папку. 
    
    Аргумент privat_page:
    
    - если True то сохранение в эксель информации из каждого файла на отдельный лист
    - False (по-умолчанию) сохранение единого датасета на 1 лист.
    
    '''
    
    # Чтение файла
    with open(file_name) as file:
        real_file = json.load(file)
    
    global_dict = {} # общий словарь для преобразования в датафрейм
    
    # Создаем пустые вспомогательные списки для заполнения словарей
    item_list = []
    SalePriceBeforePromo_list = []
    SalePriceTimePromo_list = []
    DatePriceBeforePromo_list = []
    ObjCode_list = []
    DiscountType_list = []
    DiscountValue_list = []
    
    ''' Блок Information '''

    # Проход по Goodlist
    work_dicts = real_file['Information']['GoodsLists'] # Для сокращения 
    # Количество GoodsLists в блоке Information
    cnt_goodlists = real_file['Information']['QtyGoodsLists']
    for goodlists_value in range(cnt_goodlists):

        ''' Блок Prices, ObjCode(StoreCode) '''

        # Проход по Prices
        for prices_value in  work_dicts[goodlists_value]['Prices']: # 2-й for
            #pprint(i['StoreCode'])
            # Проход по данным в Data
            for data_value in prices_value['Data']:
                # Заполняем списки значениями из даты
                item_list.append(data_value[0])
                SalePriceBeforePromo_list.append(data_value[1])
                SalePriceTimePromo_list.append(data_value[2])
                DatePriceBeforePromo_list.append(data_value[3])
                # Заполняем список значениями для каждых данных из Data для каждого Prices
                ObjCode_list.append(prices_value['StoreCode'])

            global_dict['Item'] = item_list
            global_dict['SalePriceBeforePromo'] = SalePriceBeforePromo_list
            global_dict['SalePriceTimePromo'] = SalePriceTimePromo_list
            global_dict['DatePriceBeforePromo'] = DatePriceBeforePromo_list
            global_dict['StoreCode'] = ObjCode_list # В файле это поле названо ObjCode

        ''' Блок DiscountType, DiscountValue, Value, FirstValue LessOrEqual '''
        # Следующие значения одинаковые для каждого блока GoodsLists в блоке Information

        # DiscountType, DiscountValue
        DiscountType_list.append(work_dicts[goodlists_value]['DiscountType'])
        DiscountValue_list.append(work_dicts[goodlists_value]['DiscountValue'])

        global_dict['DiscountType'] = DiscountType_list*len(item_list) # По длине значений их item
        global_dict['DiscountValue'] = DiscountValue_list*len(item_list)
        
        DiscountType_list = []
        DiscountValue_list = []

        #Value, FirstValue LessOrEqual
        for value in ['Value', 'FirstValue', 'Operator']:
            name_list = list(value+'_list')
            name_list = []
            # При анализе было выявлено, что эти поля могут отсутствовать. 
            try:
                name_list.append(work_dicts[goodlists_value]['PriceOptions'][0][value])
            except:
                name_list.append('None')
            global_dict[value] = name_list*len(item_list)
            
        
    
    # Перевод словаря в датафрейм
    df = pd.DataFrame(global_dict)
    
    ''' Блок GeneralInfo. '''
    for value in ['DateBegin', 'DateEnd', 'PWCcode']:
        
        df[value] = real_file['GeneralInfo'][value]
    # Название файла
    df['File'] = file_name
    
    
    ''' Подготовка датасета. Опции. '''
    # Косметика на соответствие ТЗ:
    
    df = df[['Item', 'SalePriceBeforePromo', 'SalePriceTimePromo',
       'DatePriceBeforePromo', 'StoreCode', 'DiscountType', 'DiscountValue',
       'DateBegin', 'DateEnd', 'PWCcode', 'Value', 'FirstValue', 'Operator', 
       'File']]
    
    return df
    
```

```python
def save_to_xls(file_list):
    it = 1
    for file_name in file_list:
        if it == 1:
            resalt_df = (JSON_pars(file_name))
            it+=1
        df = JSON_pars(file_name)
        # Соединение датафрема
        resalt_df = pd.concat([resalt_df, df])
        
    # Обновление индексов    
    resalt_df = resalt_df.reset_index()
    resalt_df.drop('index', axis=1, inplace=True)
    
    resalt_df = resalt_df.fillna('None')
    # Сохранение
    writer = pd.ExcelWriter('result.xlsx')

    resalt_df.to_excel(writer)

    writer.save()
    
    return resalt_df #, "Файл сохранен"
```
